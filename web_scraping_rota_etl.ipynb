```python
# importando biblioteca
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# definindo url
url = 'https://en.wikipedia.org/wiki/List_of_supermarket_chains'
```

```python
# fazendo requisição http com requests
response = requests.get(url)
```

```python
# analisando conteudo 1
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
# analisando conteudo 2
tabela_multinational = soup.find('table', attrs={'class': "wikitable"})
tabela_multinational
```

```python
# extraindo linhas e colunas
data = []
headers = []

# cabeçalhos
row = tabela_multinational.find('tr')
for tr in row.find_all('tr'):
  headers.append(tr.text.strip())

# dados
for row in tabela_multinational.find_all('tr')[1:]:
  columns = row.find_all('td')
  row_data = [col.text.strip() for col in columns]
  data.append(row_data)
```

```python
# criando dataframe
mercados = pd.DataFrame(data, columns=['Company', 'Headquarters', 'Served countries (besides the headquarters)', 'Map', 'Number of locations', 'Number of employees'])
mercados.head()
```

```python
# salvando dados
mercados.to_csv('/content/drive/MyDrive/mercados.csv', index=False)
print('Exportado pro drive com sucesso')
```
